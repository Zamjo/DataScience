```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
col=['sepal_length','sepal_width','petal_length','petal_width','variety']
iris=pd.read_csv("iris.csv",names=col)

print("shape:",iris.shape)
print("\nFirst five rows")
print(iris.head())
print("\nLast five rows")
print(iris.tail())
print("\nSize:",iris.size)
print("\nno of samples available for each type") 
print(iris["variety"].value_counts())
print("\n",iris.describe())
```

```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, kind ='kde',hue="species")
sns.pairplot(iris, kind ='scatter',hue="species",markers=['o','*','s'])
sns.pairplot(iris, kind ='hist',hue="species")
sns.pairplot(iris, kind ='reg',hue="species",markers=['o','*','s'])

```
